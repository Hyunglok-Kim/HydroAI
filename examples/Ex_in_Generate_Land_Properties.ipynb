```python
# 1. Select product
product = 'SMAP'

# 2. Select resolution
resolution = 'M36'  # ['M01', 'M03', 'M09', 'M36']; 1km/3km/9km/36km

# 3. Select variables
variables = ['bulk', 'sand', 'clay', 'waterfrac', 'roughness', 'UrbanFraction', 'dist2water', 'DEM', 'DEMSLP', 'dominantIGBP']
data_fp = fp_dict[product]

or_lon, or_lat, data_dict = hLP.SMAP_Land_Properties(data_fp, resolution, variables)
```

```python
# check the data
target = data_dict['bulk']
hPlot.plot_map(or_lon, or_lat, target, 0, np.nanmax(target), 'jet_r')
target = data_dict['dominantIGBP']
hPlot.plot_map(or_lon, or_lat, target, np.nanmin(target), np.nanmax(target), 'jet_r')
```
