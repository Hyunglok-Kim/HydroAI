```python
# set the data folder paths
fp_dict = {}
fp_dict['SMAP'] = cpuserver_data_FP + '/Land_properties/SMAP/'
```

```python
# land parameters from the SMAP anc data
data_fp = fp_dict['SMAP']

smap_anc_list = hData.get_file_list(data_fp, 'float32', recursive=True, filter_strs=None)
smap_anc_list = smap_anc_list + hData.get_file_list(data_fp, 'uint8', recursive=True, filter_strs=None)

# select resolution
resolutions = ['M36']     # ['M01', 'M03', 'M09', 'M36']; 1km/3km/9km/36km

# select variables
# ['DEMSLP', 'DEMSLPSTD','DEM', 'bulk', 'clay', 'sand', 'waterfrac', 'roughness', 'UrbanFraction', 'dist2water']
# some varialbes are not available for a certain resolution
variables   = ['bulk','sand', 'clay', 'waterfrac', 'roughness', 'UrbanFraction', 'dist2water', 'DEM', 'DEMSLP']

data_dict = {}
```

```python
# Process each file in the list
for file_path in smap_anc_list:
    # Determine the resolution and variable based on the file name
    resolution = None
    variable = None
    
    for res in resolutions:
        if res in file_path:
            resolution = res
            break
    
    for var in variables:
        if var in file_path:
            variable = var
            break
    
    # Determine dtype based on file extension
    if file_path.endswith('.float32'):
        dtype = np.float32
    elif file_path.endswith('.uint8'):
        dtype = np.uint8
    else:
        raise ValueError(f"Unknown file extension for {file_path}")
    
    # Read the data from the file
    data = np.fromfile(file_path, dtype=dtype)
    
    # Create a key for the dictionary
    dict_key = f"{variable}_{resolution}"
    
    # Store the data in the dictionary
    data_dict[dict_key] = data
    
    # Print some details for verification
    print(f"File: {file_path}")
    print(f"Key: {dict_key}, Data Type: {dtype}")
    print(f"Data Shape: {data.shape}, First few values: {data[:5]}")
```
