### AMSR-E/AMSR2 data [https://nsidc.org/data/au_dysno/versions/1]
AMSR-E/AMSR2 Unified Level-3 (L3) data set provides daily estimates of Snow Water Equivalent (SWE). SWE was derived from brightness temperature measurements acquired by the Advanced Microwave Scanning Radiometer 2 (AMSR2) instrument on board the JAXA GCOM-W1 satellite. The SWE data is rendered to an azimuthal 25 km Equal-Area Scalable Earth Grid (EASE-Grid) for both the Northern and Southern Hemisphere. Note: This data set uses JAXA AMSR2 Level-1R (L1R) input brightness temperatures that are calibrated, or unified, across the JAXA AMSR-E and JAXA AMSR2 L1R products.

```python
# Open the HDF5 file
file_path = '/Users/hyunglokkim/Insync/hkim@geol.sc.edu/Google_Drive/python_modules/HydroAI/examples/ex.he5'
with h5py.File(file_path, 'r') as file:
    # Print all the root level objects
    print("Root level objects in the HDF5 file:")
    for name in file:
        print(name)

    # Assuming you know the structure, access specific datasets
    # For example, if there's a dataset called 'Dataset1'
    if 'Dataset1' in file:
        dataset = file['Dataset1']
        print("Dataset1:")
        print(dataset[:])  # Print the data in Dataset1

    # To explore the structure further, recursively print all groups and datasets
    def print_structure(name, obj):
        if isinstance(obj, h5py.Dataset):
            print(f"Dataset: {name}, shape: {obj.shape}, dtype: {obj.dtype}")
        elif isinstance(obj, h5py.Group):
            print(f"Group: {name}")

    file.visititems(print_structure)
```

```python
with h5py.File(file_path, 'r') as file:
    # Extract the Northern Hemisphere Daily data
    flags_N = file['HDFEOS/GRIDS/Northern Hemisphere/Data Fields/Flags_NorthernDaily'][:]
    SWE_N = file['HDFEOS/GRIDS/Northern Hemisphere/Data Fields/SWE_NorthernDaily'][:].astype(float)
    lat_N = file['HDFEOS/GRIDS/Northern Hemisphere/lat'][:]
    lon_N = file['HDFEOS/GRIDS/Northern Hemisphere/lon'][:]
    
    # Extract the Southern Hemisphere Daily data
    flags_S = file['HDFEOS/GRIDS/Southern Hemisphere/Data Fields/Flags_SouthernDaily'][:]
    SWE_S   = file['HDFEOS/GRIDS/Southern Hemisphere/Data Fields/SWE_SouthernDaily'][:].astype(float)
    lat_S = file['HDFEOS/GRIDS/Southern Hemisphere/lat'][:]
    lon_S = file['HDFEOS/GRIDS/Southern Hemisphere/lon'][:]

# Function to replace inf/-inf with NaN in data
def replace_inf_with_nan(lat, lon, data):
    inf_mask = np.isinf(lat) | np.isinf(lon)
    data[inf_mask] = np.nan
    return data
```

```python
# Apply the function to both hemispheres
SWE_N_m = replace_inf_with_nan(lat_N, lon_N, SWE_N)
SWE_N_m[(SWE_N_m<0) | (SWE_N_m>20)] = np.nan
lat_N_m = replace_inf_with_nan(lat_N, lon_N, lat_N)
lon_N_m = replace_inf_with_nan(lat_N, lon_N, lon_N)

SWE_S_m = replace_inf_with_nan(lat_S, lon_S, SWE_S)
SWE_S_m[(SWE_S_m<0) | (SWE_S_m>20)] = np.nan
lat_S_m = replace_inf_with_nan(lat_S, lon_S, lat_S)
lon_S_m = replace_inf_with_nan(lat_S, lon_S, lon_S)
```

```python
domain_lon = longitude
domain_lat = latitude

Resampled_SWE_N = hData.Resampling(domain_lon, domain_lat, lon_N_m, lat_N_m, SWE_N_m, sampling_method='nearest', agg_method='mean', mag_factor=2)
Resampled_SWE_S = hData.Resampling(domain_lon, domain_lat, lon_S_m, lat_S_m, SWE_S_m, sampling_method='nearest', agg_method='mean', mag_factor=2)
```

Masked data if one of the following conditions was met (Kim et al. (2023)):
1) SM was smaller than 0 cm3cm− 3 higher than 0.60 cm3cm− 3
2) the Tb-RMSE flag values were larger than 5 K.
3) or the science flag was greater than one,
